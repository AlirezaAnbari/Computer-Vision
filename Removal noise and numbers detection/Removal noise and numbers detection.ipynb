```python
import numpy as np
from matplotlib import pyplot as plt

import cv2 
```

```python
# Load our image
original_image = cv2.imread("adad.jpg", cv2.IMREAD_GRAYSCALE)
original_image_1 = cv2.imread("adad.jpg")

plt.figure(figsize=[10,3])
plt.subplot(121); plt.imshow(original_image, cmap='gray'); plt.title('GrayScale Image');
plt.subplot(122); plt.imshow(original_image_1, cmap='gray'); plt.title('Image');
```

```python
#Noise removal
noise_removed_image = cv2.GaussianBlur(original_image, (9,9), 0)# or cv2.GaussianBlur(original_image,(5,5),0)

_,output1 = cv2.threshold(original_image,127,255,cv2.THRESH_BINARY)
output2 = cv2.adaptiveThreshold(original_image,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
output3 = cv2.adaptiveThreshold(original_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)

_,output4 = cv2.threshold(noise_removed_image,127,255,cv2.THRESH_BINARY)
output5 = cv2.adaptiveThreshold(noise_removed_image,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
output6 = cv2.adaptiveThreshold(noise_removed_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)


plt.figure(figsize=[15,15])
plt.subplot(331);plt.imshow(original_image, cmap='gray');plt.title("Original");
plt.subplot(332);plt.imshow(noise_removed_image, cmap='gray');plt.title("Noise removed");

plt.subplot(334);plt.imshow(output1, cmap='gray');plt.title("Global Thresholding (v = 127)");
plt.subplot(335);plt.imshow(output2, cmap='gray');plt.title("Adaptive Mean Thresholding");
plt.subplot(336);plt.imshow(output3, cmap='gray');plt.title("Adaptive Gaussian Thresholding");

plt.subplot(337);plt.imshow(output4, cmap='gray');plt.title("Global Thresholding + noise removal");
plt.subplot(338);plt.imshow(output5, cmap='gray');plt.title("Adaptive Mean + noise removal");
plt.subplot(339);plt.imshow(output6, cmap='gray');plt.title("Adaptive Gaussian + noise removal");
```

```python
image = output6.copy()
image = cv2.bitwise_not(image)   # Swap The white and black

plt.figure(figsize=[11,4])
plt.subplot(121); plt.imshow(output6, cmap='gray');plt.title("Adaptive Gaussian + noise removal");
plt.subplot(122); plt.imshow(image, cmap='gray');plt.title("Adaptive Gaussian + noise removal");
```

## Morphology

We use this method to removing salt and pepper noise in our image

```python
kernel_1 = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5), (2, 2))

erosion_1 = cv2.erode(image, kernel_1, iterations = 1)
dilation_1 = cv2.dilate(image, kernel_1, iterations = 1)
opening_1 = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel_1)
closing_1 = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel_1)

plt.figure(figsize=[15, 12])
plt.subplot(351);plt.imshow(image, cmap='gray');plt.title("Original");

plt.subplot(352);plt.imshow(erosion_1, cmap='gray');plt.title("Erosion");
plt.subplot(353);plt.imshow(dilation_1, cmap='gray');plt.title("Dilation");

plt.subplot(354);plt.imshow(opening_1, cmap='gray');plt.title("Opening");
plt.subplot(355);plt.imshow(closing_1, cmap='gray');plt.title("Closing");
plt.tight_layout()
```

## Connected Component

```python
_, image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)    # Convert Image to binary image with threshold
num_labels, labels = cv2.connectedComponents(image, connectivity=4)
print(f"Number of components : {num_labels}\n")

# Map component labels to hue val, 0-179 is the hue range in OpenCV
label_hue = np.uint8(179*labels/np.max(labels))
blank = np.ones_like(image) * 255
labeled_img = cv2.merge([label_hue, blank, blank])

# Converting cvt to BGR
labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)

# set bg label to black
labeled_img[label_hue==0] = 0


plt.figure(figsize=[15,8])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(labeled_img, cmap='gray');plt.title("labeled image");
```

We have an extra component(Blue in the corner) and we want to remove that.

```python
kernel = np.array((
        [1, 1, 1],
        [1, 1, 1],
        [1, 1, 1]), dtype="int")
output_image = cv2.morphologyEx(image, cv2.MORPH_HITMISS, kernel)

image = cv2.dilate(output_image, kernel, iterations = 2)

_, image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)    # Convert Image to binary image with threshold
num_labels, labels = cv2.connectedComponents(image, connectivity=4)
print(f"Number of components : {num_labels}\n")

# Map component labels to hue val, 0-179 is the hue range in OpenCV
label_hue = np.uint8(179*labels/np.max(labels))
blank = np.ones_like(image) * 255
labeled_img = cv2.merge([label_hue, blank, blank])

# Converting cvt to BGR
labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)

# set bg label to black
labeled_img[label_hue==0] = 0


plt.figure(figsize=[15,8])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(labeled_img, cmap='gray');plt.title("labeled image");
```
